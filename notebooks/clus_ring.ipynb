```python
graphs = sorted(glob.glob("./flow_reweighted_both_14_365/graphs/*.graphml"))

cluster_mappings = []
for graph in graphs:
    G = ig.Graph.Read_GraphML(graph)
    clusters.convert_distance_to_flow(G)
    clus = clusters.infomap(G, assign=True)
    cluster_mappings.append({k: c for k,c in zip(G.vs['name'], clus)})
```

```python
import polars as pl
```

```python
cluster_df = pl.DataFrame(schema={'node': pl.Int32})
for i, clusdict in enumerate(cluster_mappings):
    dict_frame = {'node': clusdict.keys(), f'cluster_{i}': clusdict.values()}
    df = pl.from_dict(dict_frame).select(pl.col('*').cast(pl.Int32))
    cluster_df = cluster_df.join(df, on='node', how='outer')
```
